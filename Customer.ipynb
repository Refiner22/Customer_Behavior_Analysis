```python
# Loading the dataset
import numpy as np
import pandas as pd
df = pd.read_csv(r"C:\Users\kingzuzu\Downloads\customer_shopping_behavior.csv")
df
```

```python
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Checking if missing data or null values are present in the dataset
df.isnull().sum()
```

```python
# Imputing missing values in Review Rating column with the median rating of the product category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group
labels = ['Young Adult', 'Adult', 'Middle aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# create a column purchase_frequency_days
frequency_mapping = {'Fortnightly': 14,
                     'Weekly': 7,
                     'Monthly': 30,
                     'Quarterly': 90,
                     'Bi-Weekly': 14,
                     'Annually': 365,
                     'Every 3 Months': 90}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus

# MySQL credentials
username = "root"
password = quote_plus("sunarobalee22")   # safely encode password
host = "localhost"
port = 3306
database = "customer_behavior"
table_name = "customer"

# 1️⃣ Connect WITHOUT database to ensure it exists
engine_root = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}:{port}"
)

with engine_root.connect() as conn:
    conn.execute(text(f"CREATE DATABASE IF NOT EXISTS {database}"))
    conn.commit()

# 2️⃣ Connect TO the database
engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}:{port}/{database}",
    future=True
)

# 3️⃣ Write DataFrame to MySQL
df.to_sql(
    table_name,
    con=engine,
    if_exists="replace",
    index=False,
    method="multi"   # much faster inserts
)

print(f"✅ Data successfully loaded into table '{table_name}' in database '{database}'.")
```
